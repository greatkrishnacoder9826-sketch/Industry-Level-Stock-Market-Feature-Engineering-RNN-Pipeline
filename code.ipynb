# EMA (Exponential Moving Average)

Recent Price ko zyada importance deta hai.

```python
df["EMA_10"] = ta.trend.ema_indicator(
    close=df["Close"],
    window=10
)
df["EMA_10"]
```

```python
df["EMA_20"] = ta.trend.ema_indicator(
    close=df["Close"],
    window=20
)
df["EMA_20"]
```

# RSI  - Market overbought hai ye oversold

Batata hai

Market

Overbought hai

Ya

Oversold.

```python
df["RSI"] = ta.momentum.rsi(
    close=df["Close"],
    window=14
)
df["RSI"]
```

# MACD

Momentum Indicator hai.

Ye batata hai

Trend strong hai

Ya

Weak.

Teen values milti hain.

```python
macd = ta.trend.MACD(
    close=df["Close"]
)

df["MACD"] = macd.macd()

df["MACD_Signal"] = macd.macd_signal()

df["MACD_Histogram"] = macd.macd_diff()

macd
```

# Bollinger Bands

Agar Price

Upper Band

Touch kare

↓

Market expensive.

Agar

Lower Band

Touch kare

↓

Market cheap.

Ye volatility batata hai.

```python
bb = ta.volatility.BollingerBands(
    close=df["Close"],
    window=20,
    window_dev=2
)
```

```python
df["BB_High"] = bb.bollinger_hband()

df["BB_Low"] = bb.bollinger_lband()

df["BB_Middle"] = bb.bollinger_mavg()
```

# Daily Return

```python
df["Daily_Return"] = df["Close"].pct_change()
df["Daily_Return"]
```

# Volatility

Ye batata hai

Market

Stable hai

Ya

Highly Volatile.

```python
df["Volatility"] = df["Daily_Return"].rolling(20).std()
df["Volatility"]
```

# Feature Selection

```python
features = [

    "Open",

    "High",

    "Low",

    "Close",

    "Volume",

    "EMA_10",

    "EMA_20",

    "RSI",

    "MACD",

    "MACD_Signal",

    "MACD_Histogram",

    "BB_High",

    "BB_Low",

    "BB_Middle",

    "Daily_Return",

    "Volatility"

]

data = df[features]
```

# Scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaled_data = scaler.fit_transform(data)
scaled_data
```

# Sequence Creation
Pehle hum y.append(scaled_data[i,3]) use kar rahe the, kyunki Close 4th column tha. Ab features badh gaye hain, isliye features.index("Close") use karna zyada safe hai.

```python
sequence_lenght = 60
X = []
y = []
for i in range (sequence_lenght, len(scaled_data)):
    X.append(scaled_data[i-sequence_lenght:i])
    y.append(scaled_data[i,3])
```

```python
import numpy as np
X = np.array(X)
y = np.array(y)
```

```python
train_size = int(len(X) * 0.80)
X_train = X[:train_size]
X_test = X[train_size:]
```
